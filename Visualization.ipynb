```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({"ID" : ['The Exosrcist','The Forest','Walking the rope',
                           'The Walking dead 2','Post a Terror Attack in France',
                           'Portraits of War - Syrian Refugees in Iraq',
                           'Refugee crisis in 360°: Between agony and hunger strike',
                           'Travel around the world 1','Meditation & Relaxation','Bunnies 2',
                           'Brazilian Dance','Dancing with the Stars'],         
                   "Valence" : [3.75,3.833,6.417,3.333,3.25,2.75,2.167,7.833,7.083,7.417,6.833,6.667], 
                   "Arousal" : [6.75,6.5,7.167,6.333,3.417,3.5,3.833,3.833,3.583,3.5,6,6.2]})

plt.figure(1,figsize=(10,10))
plt.xlim((1,9))
plt.ylim((1,9))
plt.plot([5,5],[1,9], linewidth=4, color='yellow' )
plt.plot([1,9],[5,5], linewidth=4, color='yellow' )
plt.xlabel('Valence')
plt.ylabel('Arousal')
plt.title('4-class plot for videos')

plt.scatter(df['Valence'], df['Arousal'],  marker='o', color='r', s=75)

for i, row in df.iterrows():
    plt.annotate(row["ID"], xy=(row["Valence"],row["Arousal"]))

plt.grid()
plt.show()
```
